## Handle Missing Values

```python
missing_object=df.columns[(df.dtypes == 'object') & (df.isna().sum()>0)]
missing_object
```

```python
for column in missing_object:
    df[column]=df[column].fillna(df[column].mode().item())
```

```python
df.isna().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median()).apply(int)
```

```python
genre_median_bpm=df[['Fav genre','BPM']].groupby('Fav genre').median()
genre_median_bpm

## i believe bpm is dependent on the genre of music
```

```python
for genre in df['Fav genre'].unique():
    mask = df['Fav genre']==genre

    median = df.loc[mask, 'BPM'].median()

    df.loc[mask, 'BPM']=df.loc[mask, 'BPM'].fillna(median)
```

```python
df.isna().sum()
```

```python
df['Timestamp']=pd.to_datetime(df['Timestamp']) #convert to datetime datatype
```

```python
for column in numerical:

    fig, ax = plt.subplots(ncols=2, figsize=(12,6))

    sns.histplot(data=df, x=column, ax=ax[0])

    sns.boxplot(data=df, y=column, ax=ax[1])

    plt.suptitle(f'{column} Distribution')
    plt.show()
```

```python
for column in categorical:
    fig, ax = plt.subplots(ncols=2, figsize=(12,5))
    
    if df[column].nunique()>4:
        explode=[.05*i if i >df[column].nunique()-4 else 0.02 for i in range(df[column].nunique())]
        rotation=75
    else:
        explode=[0.02 for i in range(df[column].nunique())]
        rotation=0

    df[column].value_counts().plot(kind='pie', startangle=90, ax=ax[0], explode=explode, autopct='%1.0f%%')
    ax[0].set_ylabel('')
    

    sns.countplot(data=df, x=column, ax=ax[1])
    
    plt.xticks(rotation=rotation)
    plt.tight_layout()
    plt.show()

    
```

```python
def classify(age):
    if age < 18:
        return 'Child'
    elif age<29:
        return 'Youth'
    elif age<59:
        return 'Adult'
    else:
        return 'Senior'
```

```python
df['Age Category'] = df['Age'].apply(classify)
```
